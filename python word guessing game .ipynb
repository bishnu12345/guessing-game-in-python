```python
#word guessing game in python 
import random
from tkinter import *

# (Lower & upper limit)
MAX = 10
MIN = 1



class Application(Frame):

    __slots__ = "number", "tries", "question", "ans", "qlabel", "list_done"

    def __init__(self, master):
        
        Frame.__init__(self, master)

        master.minsize(width=500, height=200)

        self.list_done = []

        self.qlabel = Label(self)

        self.grid()

        self.reset()

    def set_game(self):
        qno = self.number

        self.list_done.append(qno)

        if self.number == 1:
            self.question = "what is your name ."
            self.ans = "deepa"

        elif self.number == 2:
            self.question = "who is your best  footballer in world?"
            self.ans = "ronaldo"

        elif self.number == 3:
            self.question = "The greatest leader in the world?"
            self.ans = "prachanda"

        elif self.number == 4:
            self.question = "which is best programming langauge ?"
            self.ans = "python"

        elif self.number == 5:
            self.question = "what is your college name "
            self.ans ="nepal"

        elif self.number == 6:
            self.question = "What is highest peak in the world  ?"
            self.ans ="everest"


        elif self.number == 7:
            self.question = "where in gautam buddha  was born?"
            self.ans ="lumbini"


        elif self.number == 8:
            self.question = "Who is your best  singer nepal?"
            self.ans ="nirmala"


        elif self.number == 9:
            self.question = "Which is the biggest coutry in world?"
            self.ans ="japan"

        else:
            self.question = "He is known as the best crickter in world ,who is  he?"
            self.ans = "paras"

    def create_widgets(self):
        
        if self.qlabel is not None:
            self.qlabel.grid_forget()

        self.qlabel = Label(self, text = self.question)

        self.qlabel.grid(row=0, column=0, columnspan=2, sticky=W)


        Label(self,
              text="Try to guess the answer"
              ).grid(row=1, column=0, columnspan=2, sticky=W)

        Label(self,
              text="Rounds done : 0"
              ).grid(row=0, column=2, columnspan=1, sticky=W)

        Label(self,
              text="Guess"
              ).grid(row=2, column=0, sticky=W)


       
        self.guess_ent = Entry(self)
        self.guess_ent.grid(row=2, column=1, sticky=W)

        Button(self,
               text="Enter",
               command=self.get_guess
               ).grid(row=2, column=2, columnspan=4, sticky=W)
        Button(self,
               text="Reset",
               command=self.reset
               ).grid(row=1, column=2, columnspan=1, sticky=W)

        self.results_txt = Text(self, width=40, height=5, wrap=WORD)
        self.results_txt.grid(row=3, column=0, columnspan=3)

    def get_guess(self):
        
        try:
            guess = str(self.guess_ent.get())
        except(ValueError):
            self.display_message("Invalid entry. Try again.")
        else:
            self.tries += 1
            if self.tries == 10:
                self.display_message("You win!!")
                self.guess_ent.grid_forget()
               
            Label(self,
                  text="Rounds done : " + str(self.tries)
                  ).grid(row=0, column=2, columnspan=1, sticky=W)
            self.check_guess(guess)

    def next_question(self):

        self.number = random.randrange(MIN, MAX + 1)
        while self.number in self.list_done:
            self.number = random.randrange(MIN, MAX + 1)

        self.set_game()


        self.qlabel.grid_forget()

        self.qlabel = Label(self, text=self.question)

        
        self.qlabel.grid(row=0, column=0, columnspan=2, sticky=W)


    def check_guess(self, guess):
        

        if guess == self.ans:
            self.next_question()
            self.display_message("Correct Answer")
            return

        else:
            self.display_message("Wrong Answer. Game Over!")
            self.guess_ent.grid_forget()
           
            return



    def display_message(self, message):
        
        self.results_txt.delete(0.0, END)
        self.results_txt.insert(0.0, message)

    def reset(self):
        
        # Word to be guessed by player.
        self.number = random.randrange(MIN, MAX + 1)
        while self.number in self.list_done:
            self.number = random.randrange(MIN, MAX + 1)

        self.tries = 0

        self.set_game()

        self.create_widgets()
       
    def resetgame(self):
        self.display_message("Congratulationss! Your guessed is correct. The number was " + \
                             str(self.number) + ". And it only took you " + \
                             str(self.tries) + " tries!" + " Again  Click The Reset Button To Play Again" )


def main():
    root = Tk()
    root.title("Guess The Word you want ")
    app = Application(root)
    root.mainloop()



main()
```
